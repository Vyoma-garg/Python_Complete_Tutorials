```
#Dictionaries
birthdays={'vyoma':'15 oct', 'princy':'9 march','kk':'9 june'}
while True:
  print('Enter name: ')
  name=input()  #for taking input from user
  if name=='':
     break
  if name in birthdays:
    print(birthdays[name]+' is birthday of '+ name)
  else:
    print('no info for '+name)
    print('What is his/her birthday? ')
    bday=input()
    birthdays[name]=bday
    print('updated')
```

```
#Matplot
import matplotlib.pyplot as plt

x=[1,2,3,4,5,6]
y=[2,2,3,3,4,4]
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('PLOT')
plt.show()
```

```
#line plot
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0.0,2.0,0.02)
y=1+np.cos(2*np.pi*x)
plt.plot(x,y,'--')

plt.grid()
plt.xlabel('Time axis(x)')
plt.ylabel('Volatge(mv)')
plt.title('cosine function')
plt.show()
```

```
#subplot
# linspace=linearspace
x1=np.linspace(0.0,5.0)
x2=np.linspace(0.0,2.0)

y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)

plt.subplot(2,1,1) #2 rows , 1 col, 1st subplot
plt.plot(x1,y1,'*-',color='g')
plt.xlabel('x1')
plt.ylabel('y1')
plt.title('subplot1')

plt.subplot(2,1,2) #2 rows , 1 col, 2nd subplot
plt.plot(x2,y2,'.-')
plt.xlabel('x2')
plt.ylabel('y2')
plt.title('subplot2')
plt.show()
```

```
#bar plot
x=[1,2,3,4,5]
y=[10,20,40,50,30]

label=['one','two','three','four','five']
plt.bar(x,y,tick_label=label,width=0.5,color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('barplot')
plt.show()
```

```
#histogram for plotting distribution
ages=[10,20,34,45,5,56,67,7,0,90,78,14,23,43,99]
range=(0,200)
bins=20

plt.hist(ages,bins,range,histtype='bar', color='b', rwidth=0.1)
plt.xlabel(ages)
plt.ylabel('fre')
plt.title('histogramplot')
plt.show()
```

```
#scatter plot
x=[1,2,3,4,5]
y=[10,20,40,50,30]

plt.scatter(x,y,label='stars',marker='*',s=30,color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('scatterplot')
plt.show()
```

```
#pie plot

habits=['dancing','painting','singing','travel']
slices=[3,4,5,6]
color=['r','g','b','y']
plt.pie(slices,labels=habits,colors=color,shadow=True, startangle=90,explode=(0.5,0,0,0),autopct='%1.0f%%') #explode se bahar niklta, shadow se 3d look ata
plt.legend()
plt.show()
```

```
#merging of plots
x=np.arange(0,2*np.pi,0.01)
y1=np.cos(x)
y2=np.sin(x)
plt.plot(x,y1,'--',label='Cos')
plt.plot(x,y2,'o-',label='Sin')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('cosine and sine function')
plt.show()
```
